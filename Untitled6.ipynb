```python
import cv2
import numpy as Pd
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

from google.colab import files
uploaded = files.upload()
```

```python
filename=list(uploaded.keys())[0]
image=cv2.imread(filename)
cv2_imshow(image)
```

```python
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
resized=cv2.resize(gray,(800,800))
cv2_imshow(resized)
```

```python
a=[image,gray]
for i in a:
  resized=cv2.resize(i,(500,500))
  cv2_imshow(resized)
```

```python
annotated=image.copy()
cv2.rectangle(annotated,(2000,500),(1200,1000),(0,255,0),100)
cv2.putText(annotated,'MyBox',(500,1500),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,255),2)
cv2_imshow(annotated)
```

```python
cv2.imwrite('output_image.jpg',annotated)
print("image saved as output_image_jpg")
```
